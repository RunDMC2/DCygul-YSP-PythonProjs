```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import csv

read_pop = pd.read_csv('/Users/serkanaymaz/Downloads/world_pop.csv', index_col='Country Name')

world_pop = read_pop.loc['World'].values.tolist()
world_pop = world_pop[3:-1]

years = []
for i in range(1960, 2023):
    years.append(i)

# Calculate the population growth between each 10-year interval in millions
pop_growth = [(world_pop[i] - world_pop[i-1]) * 10 for i in range(1, len(world_pop))]

plt.bar(years[::10], world_pop[::10], width=5)

# Adding population growth labels with commas and a + sign
for i in range(1, len(years[::10])):
    x_pos = years[::10][i]
    y_pos = world_pop[::10][i]
    growth = int(pop_growth[i-1])
    
    # Format the growth number with commas and add a + sign
    label = '+{:,}'.format(growth)
    
    plt.text(x_pos, y_pos, label, ha='center', va='bottom')

plt.title("World Population Growth in 10-Year Increments (1960-2020)")
plt.xlabel("Year")
plt.ylabel("World Population (in billions)")

plt.show()
```
